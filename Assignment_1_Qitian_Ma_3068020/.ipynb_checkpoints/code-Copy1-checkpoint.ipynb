# Numpy Exercises

## Ex 1

```python
def inner_product(x, y):
    x = np.arrary(x)
    y = np.arrary(y)
    assert x.shape == y.shape
    return np.dot(x, y)
```

## Ex 2

```python
def mean_absolute_error(x, y):
    x = np.arrary(x)
    y = np.arrary(y)
    assert x.shape == y.shape
    return np.mean( np.abs(y - x) )
```

## Ex 3

```python
def lead(x, n):
    assert n >= 0
    x[ :-n] = x[n: ]
    x[-n: ] = [np.nan for i in range(n)]
    return x
```

```python
def lag(x, n):
    assert n >= 0
    x[n: ] = x[ :-n]
    x[ :n] = [np.nan for i in range(n)]
    return x
```

## Ex 4

```python
def distance(X, y):
    X = np.array(X)
    y = np.array(y)
    assert X.shape[1] == y.shape[0]
    return [ np.linalg.norm(X[row, :] - y, 2) for row in range(X.shape[0]) ]
```

```python
weather = pd.read_csv('data/nycflights13/nycflights13_weather.csv', skiprows = 42)
```

```python
weather.head()
```

```python
weather['temp'] = (weather['temp'] - 32) * 5/9
weather.head()
```

```python
JFK = weather[ weather['origin'] == 'JFK' ]
JFK.head()
```

```python
JFK.groupby(['year', 'month', 'day'])['hour'].count()[JFK.groupby(['year','month', 'day'])['hour'].count() != 24]
```

```python
YMD = JFK[['year', 'month', 'day']].drop_duplicates()
YMD.head()
```

```python
hour = pd.DataFrame({'hour': range(24), 'year': 2013})
hour.head()
```

```python
YMDH = pd.merge(YMD, hour, on = 'year')
YMDH.head()
```

```python
JFK_B = pd.merge(JFK, YMDH, on = ['year', 'month', 'day', 'hour'], how = 'right').sort_values(['year', 'month', 'day', 'hour'])
```

```python
JFK_B.head(10)
```

```python
JFK_B['temp'].interpolate(method = 'linear', inplace = True)
JFK_B.head(10)
```

```python
Daily_Temp = JFK_B.groupby(['month', 'day'])['temp'].mean().reset_index()
Daily_Temp.head()
```

```python
Daily_Temp['idx'] = Daily_Temp.apply(lambda row: str(int(row['month'])) + '-' + str(int(row['day'])), axis = 1)
Daily_Temp.head()
```
